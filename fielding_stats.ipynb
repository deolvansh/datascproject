## EDA

```python
##  player , there team and no. of catches 
fig= px.bar(df,x="player",y="catches",color="team",title="Catches by Player Across Teams")
```

```python
fig.show()
```

```python
# lin chart
fig= px.line(df,x="player",y="cpm",title="Line Chart of CPM Across Players")
fig.show()
```

```python
fig= px.scatter_3d(df,x="team",y="player",z="catches",color="team",title="3d_chart")
fig.show()
```

```python

fig = px.bar(df, x="player", y="catches", color="team",title="Catches by Player")
fig.show()
```

```python
fig = px.pie(df, names="team", values="catches",
             title="Team Contribution in Catches")
fig.show()
```

```python
fig = px.scatter(df, x="matches", y="catches", color="team", size="catches",
                 title="Matches Played vs Catches Taken")
fig.show()
```

```python
fig = px.bar(df.sort_values("catches", ascending=False).head(10),
             x="catches", y="player", orientation="h", color="team",
             title="Top 10 Catchers")
fig.show()
```

```python
fig = px.box(df, x="team", y="cpm", color="team",
             title="Distribution of Catches per Match by Team")
fig.show()
```

```python
fig = px.bar(df.sort_values("cpm", ascending=False).head(10),
             x="cpm", y="player", orientation="h", color="team",
             title="Top 10 Fielders by CPM")
fig.show()
```

```python
# sunburst Graph
fig = px.sunburst(df,path=["team","player","catches"],title="Graph show the Team ,  team_player , and their catches ")
fig.show()
```
