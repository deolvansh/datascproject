# EDA

```python
#  Runs Distribution by Venue
fig = px.box(df, x="venue", y="runs_of_bat", title="Runs Distribution by Venue")
fig.show()
```

```python
#  Extras Breakdown (Pie)
fig = px.pie(df, names="extras", title="Extras Breakdown")
fig.show()
```

```python
#  Wicket Types Distribution
fig = px.histogram(df[df["wicket_type"].notna()], x="wicket_type",
                    title="Distribution of Wicket Types")
fig.show()
```

```python
#  Top Bowlers by Wickets
fig = px.bar(df[df["wicket_type"].notna()].groupby("bowler")["wicket_type"].count().reset_index(),
              x="bowler", y="wicket_type", title="Top Bowlers by Wickets")
fig.show()
```

```python
#  Top Strikers by Runs
fig = px.bar(df.groupby("striker")["runs_of_bat"].sum().reset_index(),
               x="striker", y="runs_of_bat", title="Top Batsmen by Runs")
fig.show()
```

```python
#  Runs vs Extras Scatter
fig = px.scatter(df, x="runs_of_bat", y="extras", color="innings",
                   title="Runs vs Extras per Ball")
fig.show()
```

```python
#  Violin Plot → Runs Distribution by Innings
fig = px.violin(df, x="innings", y="runs_of_bat", box=True, points="all",
                 title="Runs Distribution by Innings")
fig.show()
```

```python
#  Sunburst → Batting Team → Striker → Runs
fig = px.sunburst(df, path=["batting_team","striker"], values="runs_of_bat",
                   title="Runs Contribution by Team and Striker")
fig.show()
```

```python
#  Histogram → Distribution of Runs per Ball
fig = px.histogram(df, x="runs_of_bat", nbins=6, title="Distribution of Runs per Ball")
fig.show()
```

```python
#  Treemap → Bowling Team → Bowler → Wickets
fig = px.treemap(df[df["wicket_type"].notna()],
                  path=["bowling_team","bowler","wicket_type"],
                  title="Treemap of Wickets by Team and Bowler")
fig.show()
```

```python
#  Bar Chart → Wickets by Venue
fig = px.bar(df[df["wicket_type"].notna()].groupby("venue")["wicket_type"].count().reset_index(),
              x="venue", y="wicket_type", title="Wickets by Venue")
fig.show()
```

```python
#  Violin Plot → Extras by Bowling Team
fig = px.violin(df, x="bowling_team", y="extras", box=True, points="all",
                  title="Extras Distribution by Bowling Team")
fig.show()
```

```python
#  Histogram → Extras per Ball
fig = px.histogram(df, x="extras", title="Histogram of Extras per Ball")
fig.show()
```

```python
#  Violin Plot → Runs by Over
fig = px.violin(df, x="over", y="runs_of_bat", box=True, points="all",
                  title="Runs Distribution Across Overs")
fig.show()
```

```python
# Sunburst → Venue → Innings → Runs
fig = px.sunburst(df, path=["venue","innings"], values="runs_of_bat",
                    title="Runs by Venue and Innings")
fig.show()
```

```python
# Treemap → Stage → Batting Team → Runs
fig = px.treemap(df, path=["stage","batting_team"], values="runs_of_bat",
                   title="Runs by Stage and Team")
fig.show()
```
