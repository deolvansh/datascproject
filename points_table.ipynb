##  EDA

```python
# 1 Scatter: Matches vs Teams, colored by Wins
fig = px.scatter(df, x="team", y="matches", color="wins",
                 title="Matches Played vs Wins by Team")
fig.show()
```

```python
# 2 Line: Points by Team
fig = px.line(df, x="team", y="points",
              title="Points Trend Across Teams")
fig.show()
```

```python
# 3 Line: Matches by Team
fig = px.line(df, x="team", y="matches",
              title="Matches Played by Team")
fig.show()
```

```python
# 4 Bar: Matches vs Wins
fig = px.bar(df, x="team", y="matches", color="wins",
             title="Matches Played vs Wins")
fig.show()
```

```python
# 5 Bar: Matches vs Defeats
fig = px.bar(df, x="team", y="matches", color="defeats",
             title="Matches Played vs Defeats")
fig.show()
```

```python
# 6 Bar: Ties vs Matches
fig = px.bar(df, x="ties", y="matches", color="team",
             title="Ties vs Matches by Team")
fig.show()
```

```python
# 7 Box: Matches Distribution by Team (colored by Points)
fig = px.box(df, x="team", y="matches", color="points",
             title="Distribution of Matches by Team")
fig.show()
```

```python
# 8 Pie: Matches Count by Team
count = df.groupby("team").size().reset_index(name="count")
fig = px.pie(count, names="team", values="count",
             title="Matches Distribution by Team")
fig.show()
```

```python
# 9 Sunburst: Team Outcomes Breakdown
fig = px.sunburst(df, path=["team","matches","wins","defeats","ties","abandoned"],
                  color="team", title="Team Outcomes Breakdown")
fig.show()
```

```python
# 10 Heatmap: Wins by Team
fig = px.density_heatmap(df, x="team", y="wins",
                         title="Heatmap of Wins by Team")
fig.show()
```

```python
#  Bar: Abandoned Matches by Team
fig = px.bar(df, x="team", y="abandoned", color="team",
             title="Abandoned Matches by Team")
fig.show()
```

```python
#  Bubble: Wins vs Points
fig = px.scatter(df, x="wins", y="points", size="matches", color="team",
                 title="Wins vs Points Bubble Chart")
fig.show()
```

```python
#  Scatter: Wins vs Defeats
fig = px.scatter(df, x="wins", y="defeats", color="team",
                 title="Wins vs Defeats Comparison")
fig.show()
```

```python
#  Bar: Points Table
fig = px.bar(df.sort_values("points", ascending=False),
             x="points", y="team", orientation="h", color="nrr",
             title="Points Table by Team")
fig.show()
```

```python
#  Scatter: Points vs NRR
fig = px.scatter(df, x="points", y="nrr", color="team", size="wins",
                 title="Points vs Net Run Rate")
fig.show()
```

```python
# 3-D Scatter Plot: Team Performance
fig = px.scatter_3d(df,x="team", y="matches",z="nrr", color="team",title="3D Scatter Plot: Matches vs NRR by Team")
fig.show()
```
