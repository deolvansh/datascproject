## EDA 

```python
# Sunburst: State → City → Stadium
fig = px.sunburst(df,path=["state","city","venue_stadium"],values="capacity",color="state",title="Sunburst Representation of Stadiums by State and City")
fig.show()
```

```python
fig = px.area(df.sort_values("capacity"), x="venue_stadium", y="capacity",title="Area Chart: Stadium Capacity by Venue")
fig.show()
```

```python
#  Horizontal Bar: Stadium Capacity by City
fig= px.bar(df, x="capacity", y="city", orientation="h", color="state",title="Stadium Capacity by City")
fig.show()
```

```python
#  Treemap: State → City → Stadium
fig = px.treemap(df, path=["state","city","venue_stadium"], values="capacity",color="capacity", title="Stadiums by State and City")
fig.show()
```

```python
#  Scatter: Capacity vs City
fig = px.scatter(df, x="city", y="capacity", color="state", size="capacity",title="Stadium Capacity Distribution by City")
fig.show()
```

```python
#  Sunburst: Home Team → Stadium
fig = px.sunburst(df, path=["home_team","venue_stadium"], values="capacity",color="home_team", title="Home Teams and Their Stadiums")
fig.show()
```

```python
#  Box Plot: Capacity Distribution by State
fig = px.box(df, x="state", y="capacity", color="state",title="Capacity Distribution by State")
fig.show()
```

```python

#  Bar: Stadium Capacity by Venue
fig = px.bar(df, x="venue_stadium", y="capacity", color="home_team",title="Stadium Capacity by Venue")
fig.show()
```

```python
#  Density Heatmap: State vs Capacity
fig = px.density_heatmap(df, x="state", y="capacity",title="Heatmap of Stadium Capacity by State")
fig.show()
```

```python
# Scatter 3D: City vs State vs Capacity
fig = px.scatter_3d(df, x="city", y="state", z="capacity",color="home_team", size="capacity",title="3D Scatter Plot: City, State, and Capacity")
fig.show()
```

```python
#  Pie: Stadiums by State
count_state = df.groupby("state").size().reset_index(name="count")
fig = px.pie(count_state, names="state", values="count",title="Number of Stadiums by State")
fig.show()
```
